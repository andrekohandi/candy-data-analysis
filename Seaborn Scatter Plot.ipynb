#### Scatter Plot with Seaborn

```python
# Setup/Import the libraries

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Load data

candy_filepath = "candy.csv"
candy_data = pd.read_csv(candy_filepath, index_col='id')
```

```python
candy_data.head()
```

##### 'competitorname' contains the name of the candy bar. the next 9 columns (from 'chocolate' to 'pluribus') describe the candy. For instance, rows with chocolate candies have "Yes" in the 'chocolate' column (and candies without chocolate have "No" in the same column).
##### 'sugarpercent' provides some indication of the amount of sugar, where higher values signify higher sugar content.
##### 'pricepercent' shows the price per unit, relative to the other candies in the dataset.
##### 'winpercent' is calculated from the survey results; higher values indicate that the candy was more popular with survey respondents.

```python
# Showing the relationship between 'sugarpercent' and 'winpercent'

sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
```

```python
# Showing the relationship between 'sugarpercent' and 'winpercent' with regression line

sns.regplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
```

###### Because there is correlation between 'winpercent' and 'sugarpercent', we can draw conclusions that people more like candies with more sugar in it.

```python
# Showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'

sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'], hue=candy_data['chocolate'])
```

```python
# Color-coded scatter plot w/ regression lines

sns.lmplot(data=candy_data, x='sugarpercent', y='winpercent', hue='chocolate')
```

```python
# Showing the relationship between 'chocolate' and 'winpercent'

sns.swarmplot(x=candy_data['chocolate'],
             y=candy_data['winpercent'])
```

##### We can draw conclusion here, the chocolate candies is more popular rather than the candies without chocolate
